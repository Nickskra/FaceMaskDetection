```python
import os
import matplotlib.pyplot as plt
```

```python
img_path = "../data/Dataset"

classes = os.listdir(img_path)
counts = [len(os.listdir(os.path.join(img_path, c))) for c in classes]
```

```python
plt.figure(figsize=(12,6))
bars = plt.bar(classes, counts)
plt.title("Persebaran Dataset")

for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2, bar.get_height(), bar.get_height(),
             ha='center', va='bottom')

plt.show()
```
